```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
template2 = cv2.imread('m2.pgm', 0)
template3 = cv2.imread('m3.pgm', 0)
template4 = cv2.imread('m4.pgm', 0)
plt.figure(1, figsize=(12,12))
plt.imshow(template2)
plt.figure(2, figsize=(12,12))
plt.imshow(template3)
plt.figure(3, figsize=(12,12))
plt.imshow(template4)
```

```python
img_rgb = cv2.imread('./imgs/m4_002.jpg')
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
plt.figure(1, figsize=(12,12))
plt.imshow(img_gray)

```

```python
#height, width = template.shape[:]


res2 = cv2.matchTemplate(img_gray, template2, cv2.TM_CCOEFF_NORMED)
res3 = cv2.matchTemplate(img_gray, template3, cv2.TM_CCOEFF_NORMED)
res4 = cv2.matchTemplate(img_gray, template4, cv2.TM_CCOEFF_NORMED)


plt.figure(1, figsize=(8,8))
plt.imshow(res2, cmap='gray')
plt.figure(2, figsize=(8,8))
plt.imshow(res3, cmap='gray')
plt.figure(3, figsize=(8,8))
plt.imshow(res4, cmap='gray')
```

```python
#min_val, max_val, min_loc, max_loc
valor2 = cv2.minMaxLoc(res2)
valor3 = cv2.minMaxLoc(res3)
valor4 = cv2.minMaxLoc(res4)



print(np.sum(res2)/np.sum(res2+res3+res4))
print(np.sum(res3)/np.sum(res2+res3+res4))
print(np.sum(res4)/np.sum(res2+res3+res4))
```
